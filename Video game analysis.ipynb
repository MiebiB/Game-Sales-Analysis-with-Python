```python
import pandas as pd
import numpy as np

ts=pd.read_csv('C:/Users/USER/Downloads/VideoGamesProject-main/VideoGamesProject-main/VideoGamesSales.csv') #imports the data
df=ts[ts.duplicated()] #shows all duplicate rows
df
```

```python
ts=ts.drop_duplicates() #deletes all the duplicate rows
ts.shape[0] #new count (5909-16=5893)
```

```python
#to replace missing values
ts['Region'] = ts['Region'].replace(r'^\s*$', np.nan, regex=True) #replaces empty strings or strings with only spaces with np.nan
ts['Region'] = ts['Region'].fillna('North') # Now use fillna to replace actual missing values
ts['Region'].isnull().sum() #to verify that there are no missing values
```

```python
ts.shape[1]
```

```python
#COuntry column needs cleaning
ts.head(10) 
```

```python
ts['Country']=ts['Country'].replace({'USA':'United States'}) #replace USA with United States
ts['Country']=ts['Country'].str.title() #puts the first letters in uppercase
ts.head(10)
```

```python
ts
```

```python
#cleaning up NA_Sales data
ts['NA_Sales']=ts['NA_Sales'].replace(r'[$]', '', regex=True) #replace $ with nothing
ts['NA_Sales']=pd.to_numeric(ts['NA_Sales'], errors='coerce') #change to float numeric
Average_sales=ts['NA_Sales'].mean()
Average_sales=int(Average_sales)
Average_sales #gives 0 as average
ts['NA_Sales']=ts['NA_Sales'].fillna(Average_sales) #fill empty cells with average sale
ts
```

```python
#to rename column names
ts=ts.rename(columns={'NA_Sales':'National Sales', 'Global_Sales':'Global Sales', 'NA_Profit':'National Profit', 'Global_Profit':'Global Profit'})
ts
```

```python
#some values in National Sales are way above average and could hinder accurate analysis
sales_cap=ts['National Sales'].quantile(0.95) #means that 95% of values are less than what is shown what i execute sales_cap
#sales_cap #2.07

ts['National Sales']=np.where(ts['National Sales'] > sales_cap, sales_cap, ts['National Sales']) #to replace those big values with 2.07
ts
```

```python
# doing the same for Global sales
gsales_cap=ts['Global Sales'].quantile(0.95) 
#gsales_cap #4.04

ts['Global Sales']=np.where(ts['Global Sales'] > gsales_cap, gsales_cap, ts['Global Sales'])
ts
```

```python
# doing the same for National Profit
profit_cap=ts['National Profit'].quantile(0.95) 
#profit_cap #0.621

ts['National Profit']=np.where(ts['National Profit'] > profit_cap, profit_cap, ts['National Profit'])
ts
```

```python
# doing the same for Global Profit
gprofit_cap=ts['Global Profit'].quantile(0.95) 
#gprofit_cap #1.211

ts['Global Profit']=np.where(ts['Global Profit'] > gprofit_cap, gprofit_cap, ts['Global Profit'])
ts
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#to create bar chart of National sales by region

National_Sales=ts.groupby(['Region','Country'])['National Sales'].sum().reset_index().sort_values(by='National Sales', ascending=False)
National_Sales
plt.figure(figsize=(16,8))
sns.barplot(data=National_Sales, x='Region', y='National Sales', hue='Country')
plt.title('National Sales by region and country')
plt.xlabel('region')
plt.ylabel('National Sales')
plt.show()
```

```python
#create box plot National sales by country

plt.figure(figure=(10,6))
sns.boxplot(x='Country', y='National Sales', data=ts, hue='Genre')
plt.title('National Sales by Country')
plt.xlabel('Country', fontweight='bold')
plt.ylabel('National Sales', fontweight='bold')
plt.show()
```

```python
#Create pie chart comparison of national and global sales by country

Sales = ts.groupby(['Country'])[['National Sales', 'Global Sales']].sum().reset_index()
Country=Sales['Country']
National_Sales=Sales['National Sales']
Global_Sales=Sales['Global Sales']

fig, axs=plt.subplots(1,2,figsize=(10,7))

axs[0].pie(National_Sales, labels=Country, autopct='%1.1f%%',startangle=30)
axs[0].set_title('National Sales by Country')

axs[1].pie(Global_Sales, labels=Country, autopct='%1.1f%%',startangle=30)
axs[1].set_title('Global Sales by Country')

plt.show()
```

```python
#create line chart for year over year sales

sales_by_year = ts.groupby('Year')[['National Sales', 'Global Sales']].sum().reset_index()

plt.figure(figsize=(10,6))

plt.plot(sales_by_year['Year'], sales_by_year['National Sales'], marker='o', linestyle='-', color='b', label='National Sales') #line plot for National Sales

plt.plot(sales_by_year['Year'], sales_by_year['Global Sales'], marker='s', linestyle='--', color='g', label='Global Sales') #line plot for Global Sales

plt.title('National and Global Sales Over Years')
plt.xlabel('Year', fontweight='bold')
plt.ylabel('Sales', fontweight='bold')
plt.legend()

plt.grid(True) #add grid
plt.show()
```

```python
#to create bar chart of Total sales by region

region_sales = ts.groupby('Region')[['National Sales', 'Global Sales']].sum().sort_values('Global Sales', ascending=False)

region_sales.plot(kind='bar', figsize=(10,6), colormap='viridis')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales (in millions)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# To find the 10 genre by global sales

top_games = ts.groupby('Genre')['Global Sales'].sum().sort_values(ascending=False).head(10)

# Plot
top_games.plot(kind='barh', figsize=(10,6), color='skyblue')
plt.title('Top 10 Best-Selling Game Genre Globally')
plt.xlabel('Global Sales')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Top platforms by total sales

top_platforms = ts.groupby('Platform')['Global Sales'].sum().sort_values(ascending=False).head(10)

top_platforms.plot(kind='bar', figsize=(10,6), color='salmon')
plt.title('Top 10 Platforms by Global Sales')
plt.ylabel('Global Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Genre popularity accross regions
genre_region = ts.groupby(['Genre', 'Region'])['Global Sales'].sum().unstack().fillna(0)
#genre_region

genre_region.plot(kind='bar', figsize=(12,6), stacked=True, colormap='tab20')
plt.title('Regional Preferences by Game Genre')
plt.xlabel('Genre')
plt.ylabel('Global Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
